# Define data augmentation using Albumentation library

  *The AUGMENTATIONS sequence includes a variety of augmentation operations such as random resizing and cropping, horizontal and vertical flips, transposition, adding Gaussian noise, sharpening, embossing, rotation, and applying one of several contrast and brightness adjustments. Each operation introduces specific variations to the input images, such as changes in orientation, scale, noise, and color balance.*

```python
import albumentations as A

AUGMENTATIONS = A.Compose([
    A.RandomResizedCrop(224, 224),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.Transpose(p=0.5),
    A.GaussNoise(p=0.5),
    A.Sharpen(p=0.5),
    A.Emboss(p=0.5),
    A.Rotate(p=0.5),
    A.OneOf([A.CLAHE(p=1), A.RandomBrightnessContrast(p=1), A.RandomGamma(p=1),], p=0.5),
])
```

```python
def plot_augmented_images(folder_path, augmentations):
    files = os.listdir(folder_path)
    image_files = [file for file in files if file.endswith('.jpg')]

    selected_images = image_files[:3]

    fig, axes = plt.subplots(2, 3, figsize=(12, 8))

    for i, image_file in enumerate(selected_images):
        image_path = os.path.join(folder_path, image_file)
        image = mpimg.imread(image_path)

        augmented = augmentations(image=image)
        augmented_image = augmented['image']

        axes[0, i].imshow(image)
        axes[0, i].axis('off')
        axes[0, i].set_title('Original')

        axes[1, i].imshow(augmented_image)
        axes[1, i].axis('off')
        axes[1, i].set_title('Augmented')

    plt.show()
```

```python
# Example usage:
folder_path = "/content/valid/nowildfire/"
plot_augmented_images(folder_path, AUGMENTATIONS)
```

*The provided function, plot_augmented_images, is designed to visualize original images alongside their augmented versions. It takes a folder path containing image files and a set of augmentation transformations as input. It then selects a subset of images from the folder, applies the specified augmentations to each selected image, and displays the original and augmented images side by side for comparison. This function is useful for understanding how different augmentation techniques affect the appearance of images and can be used to assess the effectiveness of augmentation strategies in data preprocessing for machine learning tasks.*

# Define ImageDataAugmentor for preprocessing and data augmentation

```python
datagen = ImageDataAugmentor(rescale=1.0 / 255)

train_datagen = ImageDataAugmentor(
        rescale=1./255,
        augment=AUGMENTATIONS,
        preprocess_input=None)

# Create a generator for training data
train_generator = train_datagen.flow_from_directory(
    os.path.join(dataset_path, 'train'),
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary'
)

# Create a generator for validation data
validation_generator = datagen.flow_from_directory(
    os.path.join(dataset_path, 'valid'),
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary'
)

# Create a generator for testing data
test_generator = datagen.flow_from_directory(
    os.path.join(dataset_path, 'test'),
    target_size=(img_width_test, img_height_test),
    batch_size=batch_size,
    class_mode='binary'
)
```

*In this code snippet, image data generators are set up for training, validation, and testing using the ImageDataAugmentor class. These generators are essential for loading and preprocessing image data, making it ready for consumption by machine learning models. Here's an overview of what each part of the code does:*

**ImageDataAugmentor Setup:**
*Two instances of ImageDataAugmentor are created: datagen and train_datagen.
datagen is used for general data processing without augmentation, only rescaling the pixel values to a range between 0 and 1.
train_datagen is specifically configured for training data. It rescales pixel values and applies augmentation transformations defined in the AUGMENTATIONS variable, which likely contains various image augmentation techniques.
Creating Generators:
train_generator, validation_generator, and test_generator are created using the flow_from_directory method of the respective data generators.
train_generator is configured to load training data from the 'train' directory, resize images to the specified dimensions, and generate batches of data along with binary class labels.
validation_generator and test_generator are set up similarly but for validation and testing data, loading images from the 'valid' and 'test' directories respectively.*

# Define model: MobileNet
*Howard, Andrew G., et al. MobileNets: Efficient Convolutional Neural Networks for Mobile Vision Applications. 2017. DOI.org (Datacite), https://doi.org/10.48550/ARXIV.1704.04861.*

```python
from classification_models.keras import Classifiers

mobilenet, preprocess_input = Classifiers.get('mobilenet')

# build model
base_model = mobilenet(input_shape=(224,224,3), weights=None, include_top=False)
x = keras.layers.GlobalAveragePooling2D()(base_model.output)
output = keras.layers.Dense(1, activation='sigmoid')(x)
model = keras.models.Model(inputs=[base_model.input], outputs=[output])
```

# Compile the model

```python
# Compile the model
opt = keras.optimizers.Adam(learning_rate=1e-5)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['AUC','accuracy','TruePositives','TrueNegatives','FalsePositives','FalseNegatives'])
```

```python
model.summary()
```

# Define callbacks

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

# Define the callback
early_stopping = EarlyStopping(monitor='val_loss',  # Monitor
                               patience=10,          # Number of epochs with no improvement to stop training
                               verbose=1,           # Print a message when training stops early
                               mode='min',          # Monitored metric should be minimized
                               restore_best_weights=False)  # Restore the weights of the best model

#min_delta=0.001,     # Minimum change to be considered as improvement

checkpoint = ModelCheckpoint(filepath='/content/best_model.h5',
                             monitor='val_loss',
                             save_best_only=True,
                             verbose=1)

#class_weight = {0: 0.5, 1: 1}
```

# Train the model

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.n // batch_size,
    epochs=100,
    batch_size=batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.n // batch_size,
    callbacks=[early_stopping, checkpoint]
)
#class_weight = class_weight,
```

# Plot performance graphs

```python
 # Show performance graphs
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
 # Show performance graphs
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['false_positives'], label='Train FP')
plt.plot(history.history['val_false_positives'], label='Validation FP')
plt.title('False Positive')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['false_negatives'], label='Train FN')
plt.plot(history.history['val_false_negatives'], label='Validation FN')
plt.title('False Negative')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
